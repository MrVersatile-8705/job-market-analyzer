# Results Visualization

This notebook is dedicated to visualizing the results of the job market analysis conducted on over 1 million job descriptions. The visualizations will help in understanding trends, skill requirements, and salary distributions.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')

# Load the results data
results_df = pd.read_csv('../data/results/analysis_results.csv')

# Display the first few rows of the dataframe
results_df.head()
```

```python
# Visualize the distribution of salaries
plt.figure(figsize=(12, 6))
sns.histplot(results_df['salary'], bins=30, kde=True)
plt.title('Salary Distribution of Data Jobs')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the most common skills required
top_skills = results_df['skills'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_skills.values, y=top_skills.index, palette='viridis')
plt.title('Top 10 Most Common Skills in Job Descriptions')
plt.xlabel('Number of Mentions')
plt.ylabel('Skills')
plt.show()
```

```python
# Visualize salary by job role
plt.figure(figsize=(12, 6))
sns.boxplot(x='job_role', y='salary', data=results_df)
plt.title('Salary by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
# Save visualizations to files
plt.figure(figsize=(12, 6))
sns.histplot(results_df['salary'], bins=30, kde=True)
plt.title('Salary Distribution of Data Jobs')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.savefig('../data/results/salary_distribution.png')
plt.close()
```
