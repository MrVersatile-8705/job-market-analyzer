# Exploratory Data Analysis on Job Market Data

In this notebook, we will perform exploratory data analysis (EDA) on the job market data collected from various sources. The goal is to understand the distribution of job descriptions, identify key trends, and visualize important metrics.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')
```

```python
# Load the job market data
data_path = '../data/processed/job_market_data.csv'
job_data = pd.read_csv(data_path)

# Display the first few rows of the dataset
job_data.head()
```

```python
# Summary statistics of the dataset
job_data.describe(include='all')
```

```python
# Visualize the distribution of job titles
plt.figure(figsize=(12, 6))
job_data['job_title'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analyze the distribution of salaries
plt.figure(figsize=(12, 6))
sns.histplot(job_data['salary'], bins=30, kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation heatmap of numerical features
plt.figure(figsize=(10, 8))
correlation_matrix = job_data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## Conclusion

In this exploratory analysis, we have visualized key aspects of the job market data, including job title distributions, salary distributions, and correlations between various features. These insights will guide further analysis and model development.
